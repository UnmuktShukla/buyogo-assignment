```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data['children'].fillna(0, inplace=True)
data['country'].fillna('Unknown', inplace=True)
data['agent'].fillna('0', inplace=True)
data['company'].fillna('0', inplace=True)

missing_values_after = data.isnull().sum()
print(missing_values_after)
```

```python
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])
monthMap = {'January':1 , 'February' : 2 , 'March' : 3 , 'April' : 4 , 'May' : 5 , 'June' : 6 , 'July' : 7 , 'August' : 8 , 'September' : 9 , 'October' : 10 , 'November' : 11 , 'December' : 12}

data['arrival_date_month'] = data['arrival_date_month'].map(monthMap)
```

```python

mean_lead_time = data['lead_time'].mean()
median_lead_time = data['lead_time'].median()
min_lead_time = data['lead_time'].min()
max_lead_time = data['lead_time'].max()
percentiles = data['lead_time'].quantile([0.25, 0.75])


print("📊 Booking Lead Time Distribution Analysis")
print(f"Mean lead time: {mean_lead_time:.2f} days")
print(f"Median (50th percentile): {median_lead_time:.2f} days")
print(f"Most common range: {percentiles[0.25]:.2f} to {percentiles[0.75]:.2f} days")
print(f"Max lead time: {max_lead_time:.2f} days (outlier cases)")

# Identifying trend
if mean_lead_time > median_lead_time:
    print("\nTrend: A right-skewed distribution, indicating that while many bookings are made months in advance, some occur just days before arrival.")
else:
    print("\nTrend: A more uniform or normal distribution.")

print("\nThis insight helps hotels in pricing strategies and capacity planning.")
```

```python
data["text"] = data.apply(lambda row: f"Hotel: {row['hotel']}, City: {row['country']}, "
                                  f"Total Price: {row['adr']*(row['stays_in_week_nights'] + row['stays_in_weekend_nights'])}, "
                                  f"Lead Time: {row['lead_time']} days, "
                                  f"Status: {row['reservation_status']}", axis=1)

data = data.dropna(subset=["text"])

# Display sample data
data[["text"]].head()
```

```python
#generating vector embeddings
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(data["text"].tolist() , normalize_embeddings=True)
embeddings = np.array(embeddings)

print("Embedding shape: " , embeddings.shape)
```

```python
import faiss
```

```python
d = embeddings.shape[1]
index = faiss.IndexFlatL2(d)
index.add(embeddings)
faiss.write_index(index, "faiss_index.bin")
print("FAISS Index Created! Number of entries:", index.ntotal)
```
